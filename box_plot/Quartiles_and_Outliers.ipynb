### üìö –Ü–º–ø–æ—Ä—Ç –±—ñ–±–ª—ñ–æ—Ç–µ–∫

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### üìå –†–∞—Ö—É—î–º–æ –∫–≤–∞—Ä—Ç–∏–ª—ñ (quantiles)

**–ö–≤–∞—Ä—Ç–∏–ª—ñ** –ø–æ–∑–Ω–∞—á–∞—é—Ç—å—Å—è —è–∫ Q1 (–Ω–∏–∂–Ω—ñ–π –∫–≤–∞—Ä—Ç–∏–ª—å), Q2 (–º–µ–¥—ñ–∞–Ω–∞) —ñ Q3 (–≤–µ—Ä—Ö–Ω—ñ–π –∫–≤–∞—Ä—Ç–∏–ª—å). –ù–∏–∂–Ω—ñ–π –∫–≤–∞—Ä—Ç–∏–ª—å –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î 25-–º—É –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—é, –∞ –≤–µ—Ä—Ö–Ω—ñ–π –∫–≤–∞—Ä—Ç–∏–ª—å –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î 75-–º—É –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—é, —Ç–æ–º—É —î –ø–æ–Ω—è—Ç—Ç—è **–º—ñ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ —Ä–æ–∑–º–∞—Ö—É**:

$$ ùêºùëÑùëÖ=ùëÑ3‚àíùëÑ1 $$

```python
# –î–∞–Ω—ñ
data = [47, 12, 85, 23, 56, 91, 34, 72, 18, 63]

# –†–∞—Ö—É—î–º–æ –∫–≤–∞–Ω—Ç–∏–ª—ñ and IQR –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º NumPy
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

```python
sns.set_palette("pastel")

# –ü–æ–±—É–¥–æ–≤—É—î–º–æ –≥—Ä–∞—Ñ—ñ–∫
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, color="#D8BFD8", showfliers=False)

# –î–æ–¥–∞—î–º–æ –∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ñ –ª—ñ–Ω—ñ—ó –¥–ª—è –≤—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—ó
plt.axhline(y=q1, color="orange", linestyle="--", label="Q1")
plt.axhline(y=q2, color="blue", linestyle="--", label="Q2 (Median)")
plt.axhline(y=q3, color="green", linestyle="--", label="Q3")

# –ó–∞–¥–∞—î–º–æ –æ—Å—ñ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–æ–∫
plt.xlabel("data")
plt.ylabel("values")
plt.title("Boxplot")

# –î–æ–¥–∞—î–º–æ –ª–µ–≥–µ–Ω—Ç—É
plt.legend()

# –ë—É–¥—É—î–º–æ –≥—Ä–∞—Ñ—ñ–∫
plt.show()
```

### üìå –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—É (outliers)
**–í–∏–∫–∏–¥–∏** - —É —Å—Ç–∞—Ç–∏—Å—Ç–∏—Ü—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤–∏–º—ñ—Ä—é–≤–∞–Ω–Ω—è, —è–∫–∏–π –≤–∏–¥—ñ–ª—è—î—Ç—å—Å—è —ñ–∑ –∑–∞–≥–∞–ª—å–Ω–æ—ó –≤–∏–±—ñ—Ä–∫–∏.

```python
# –ì–µ–Ω–µ—Ä—É—î–º–æ —Ä–∞–Ω–¥–æ–º–Ω—ñ –¥–∞–Ω—ñ –∑ –≤–∏–∫–∏–¥–∞–º–∏
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(8, 1, 10)])

# –ó–∞–¥–∞—î–º–æ –∫–æ–ª—ñ—Ä –≥—Ä–∞—Ñ—ñ–∫—É
pastel_green_palette = sns.color_palette(["#a1e579"])

# –°—Ç–≤–æ—Ä—é—î–º–æ box plot
sns.set(style="white", palette=pastel_green_palette)
plt.figure(figsize=(7, 3))

sns.boxplot(x=data)

# –î–æ–¥–∞—î–º–æ –∑–∞–≥–æ–ª–æ–≤–æ–∫ —ñ –º—ñ—Ç–∫–∏
plt.title("Boxplot with Outliers")
plt.xlabel("data")

# –í–∏–≤–æ–¥–∏–º–æ –≥—Ä–∞—Ñ—ñ–∫ –Ω–∞ –µ–∫—Ä–∞–Ω
plt.show()
```
