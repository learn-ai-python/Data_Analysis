### üìö –Ü–º–ø–æ—Ä—Ç –±—ñ–±–ª—ñ–æ—Ç–µ–∫

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### üå∑ –ó—á–∏—Ç—É–≤–∞–Ω–Ω—è –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö

```python
df = sns.load_dataset('iris')
df.head()
```

### üìà –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é box plot

```python
sns.boxplot(x=df["sepal_length"],  # —Ü—ñ–ª—å–æ–≤–∞ –∫–æ–ª–æ–Ω–∫–∞
            palette=["#e0a6ff"],  # –∫–æ–ª—ñ—Ä –≥—Ä–∞—Ñ—ñ–∫–∞
            showmeans=True)  # –≤—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è

plt.show()
```
